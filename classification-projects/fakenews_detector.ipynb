```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
true = pd.read_csv('True.csv')
fake = pd.read_csv('Fake.csv')
```

```python
true['Label'] = 1
fake['Label'] = 0
```

```python
news = pd.concat([true, fake], axis = 0)
news.head()
```

```python
news.isnull().sum()
```

```python
news = news.drop(['title', 'subject', 'date'], axis =1)
```

```python
# Reshuffling
news = news.sample(frac = 1)
news.head()
```

```python
# Resetting the index to be in order
news = news.reset_index()
news.head()
```

```python
news = news.drop(['index'], axis = 1)
news.head()
```

```python
import re
import string
```

```python
def wordopt(text):
    text = text.lower()
    # Remove URLs first
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # Remove HTML tags
    text = re.sub(r'<.*?>+', '', text)
    # Remove punctuation
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    # Remove digits
    text = re.sub(r'\d+', '', text)
    # Remove non-alphanumeric characters (except space)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
X = news['text']
y = news['Label']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
# Creating vectors
xv_train = vectorization.fit_transform(X_train)
xv_test = vectorization.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(xv_train, y_train)
```

```python
pred_lr = lr.predict(xv_test)
lr.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(xv_train, y_train)

pred_dt = dt.predict(xv_test)
dt.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(xv_train, y_train)

pred_gb = gb.predict(xv_test)
gb.score(xv_test, y_test)

print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(xv_train, y_train)

pred_rf = rf.predict(xv_test)
rf.score(xv_test, y_test)

print(classification_report(y_test, pred_rf))
```

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"
```

```python
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = lr.predict(new_xv_test)
    pred_DT = dt.predict(new_xv_test)
    pred_GB = gb.predict(new_xv_test)
    pred_RF = rf.predict(new_xv_test)
    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGB Prediction: {} \nRF Prediction: {}".format(
        output_lable(pred_LR[0]), output_lable(pred_DT[0]), output_lable(pred_GB[0]), output_lable(pred_RF[0])
        )
    )
```
